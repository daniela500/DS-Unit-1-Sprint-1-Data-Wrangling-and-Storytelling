### there is usually more than 1 way to do things

```python
# this returns pandas version
pd.__version__
```

```python
int_rate = ' 14.47%'
```

```python
# .strip() removes characters from begining and end if they are found, if character isnot defined it assumes whitespace (' ')
int_rate = int_rate.strip('%')
int_rate.strip()
```

```python
# chaining of methods
int_rate = ' 14.47%'
int_rate.strip().strip('%')
```

```python
int_rate.replace('%','').replace(' ','')
```

```python
type(int_rate.strip().strip('%'))
```

```python
# cast a string to a float
float('14.45')
type(float('14.45')) 
```

```python
# will write all the above in a function
def int_rate_to_float(int_rate_str):
  return float(int_rate_str.strip().strip('%'))

int_rate_to_float(int_rate)  
# type(int_rate_to_float(int_rate))  
```

Apply the function to the `int_rate` column

```python
# apply() loops through all the values in the column and applies the specified function to them
 df['int_rate'].apply(int_rate_to_float)
```

```python
df1['int_rate'] = df['int_rate']
df1['int_rate_float'] = df['int_rate'].apply(int_rate_to_float)
df1.head()
```

Clean the title and handle missing values

```python
example = ['owner', 'Suprevisor', ' Project Manager', np.NaN]
print(example)
```

```python
# title() makes Title casing
# isinstance() checks the type of the passed value and returns True or False

def clean_title(title):
  if isinstance(title, str): 
    return title.strip().title()
  else: 
    return 'Unknown'
# to call the function with param  
clean_title(' owner')  
```

```python
# to loop through the items of a List 
 [clean_title(item) for item in example]
# the result is returned as a list again
```

```python
# second way to loop the items of a list
for item in example:
  print(clean_title(item)) 
# the result is printing each value while looping
```

```python
df1['emp_title'] = df['emp_title'].apply(clean_title)
len(df1['emp_title'].value_counts())
```
